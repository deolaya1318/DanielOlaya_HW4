# Estudio de la variación entre la gravedad $g_y$ y el ángulo polar $\theta$

A continuación se ha graficado la relación entre la gravedad $g_y$ y el ángulo polar $\theta$.

```python
fig, xy = plt.subplots(1, 1, figsize=(8, 5))
xy.scatter(theta, g_y, color = 'b')
xy.set_title(r"Gravedad versus $\theta$")
xy.set_xlabel(r"$\theta$")
xy.set_ylabel("$g_y$")
```

Para analizar esta relación se ha propuesto que la variación de la gravedad con el ángulo $\theta$ es parametrizada a partir de la siguiente función:

$F(g) = 1 - g/\langle g\rangle$

A continuación se define esta función y se realiza la gráfica resultante.

```python
gpromedio = sum(g_y)/len(g_y)
def F(g_y):
    return 1 - g_y/gpromedio
```

```python
fig, xy= plt.subplots(1, 1, figsize=(8, 5))
xy.scatter(theta, F(g_y),color ='r')
xy.set_title(r"$F(g_y)$ versus $\theta$")
xy.set_xlabel(r"$\theta$")
xy.set_ylabel("$F(g_y)$")
```

Por último queremos saber cual de los siguiente modelos puede representar de una mejor manera esta variación de la gravedad con respecto al ángulo $\theta$. Para ello debemos crear un archivo que será la entrada a los respectivos códigos que realizan el ajuste de mínimos cuadrados para cada modelo.

```python
x = linspace(0,180,900)
arch = open('datos_gravedad.dat','w')
for i in range(len(theta)):
    arch.write("%f %f \n" %(theta[i], F(g_y[i])))
arch.close()
```

## Modelo 1

De acuerdo a este modelo la función parametrizada es de la siguiente forma:
$F = a_0 + a_1*\cos(2\theta)$

Usando el código fuente "hipotesis1.c" (guardado en el repositorio) se han encontrado los coeficientes $a_0$ y $a_1$. Al igual que el anterior código, las respectivas explicaciones se encuentran dentro del código. A partir de estos coeficientes se ha gráficado la respuesta que predice el modelo (línea verde continua) contra la variación de los datos experimentales (puntos rojos discontinuos).

```python
!cc hipotesis1.c -lm
!./a.out
```

```python
coeficientes1 = np.loadtxt("parametros_hipotesis1.dat")
print "a_0 =",coeficientes1[0], "a_1 =",coeficientes1[1]
```

```python
def H1(x):
    return coeficientes1[0]+coeficientes1[1]*cos(2*x*pi/180.0)
fig, xy = plt.subplots(1, 1, figsize=(8, 5))
xy.scatter(theta, F(g_y), color = 'r', label = "Ajuste a datos")
xy.plot(x, H1(x), color = 'g', label= "Hipotesis 1")
xy.set_title(r"$F(\theta)$ versus $\theta$")
xy.set_xlabel(r"$\theta$")
xy.set_ylabel(r"$F(\theta)$")
handles, labels = xy.get_legend_handles_labels()
xy.legend(handles, labels, loc = 3)
```

## Modelo 2

De acuerdo a este modelo la función parametrizada es de la siguiente forma:
$F = b_0 + b_1*\theta + b_2*\theta²$

Usando el código fuente "hipotesis2.c" (guardado en el repositorio) se han encontrado los coeficientes $b_0$, $b_1$ y $b_2$. Al igual que los anteriores códigos, las respectivas explicaciones se encuentran dentro del código. A partir de estos coeficientes se ha gráficado la respuesta que predice el modelo (línea verde continua) contra la variación de los datos experimentales (puntos rojos discontinuos).

```python
!cc hipotesis2.c
!./a.out
```

```python
coeficientes2 = np.loadtxt("parametros_hipotesis2.dat")
print "b_0 =",coeficientes2[0], "b_1 =",coeficientes2[1], "b_2 =", coeficientes2[2]
```

```python
def H2(x):
    return coeficientes2[0]+coeficientes2[1]*x+coeficientes2[2]*x**2
fig, xy = plt.subplots(1, 1, figsize=(8, 5))
xy.scatter(theta, F(g_y), color = 'r', label = "Ajuste a datos")
xy.plot(x, H2(x), color = 'g', label= "Hipotesis 2")
xy.set_title(r"$F(\theta)$ versus $\theta$")
xy.set_xlabel(r"$\theta$")
xy.set_ylabel(r"$F(\theta)$")
handles, labels = xy.get_legend_handles_labels()
xy.legend(handles, labels, loc = 3)
```

## Comparación entre los modelos

Al comparar estas dos gráficas es de notar que en todo el rango de $/theta$, la primera hipotesis es la que mejor se ajusta a los datos experimentales. Solo en una franja pequeña cercana a los 70° y 90° se nota que la hipotesis 2 puede predecir de una mejor manera el comportamiento físico de la gravedad con respecto a este ángulo. En contraste para ángulos muy pequeños y muy grandes, la primera hipotesis se ajusta mucho mejor que la segunda hipotesis en un mayor rango, por lo cual se concluye que el mejor modelo para explicar la variación de la gravedad es la primera hipotesis.

NOTA: Las unidades en todas las gráficas de este Notebook son: para el ángulo $\theta$ son [grados] y para la gravedad $g_y$ son [longitud/tiempo²]. La función $F(g)$ es adimensional.
